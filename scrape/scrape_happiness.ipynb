```python
# Import libraries for scraping

from __future__ import print_function, division
import requests
import re
import pandas as pd
from bs4 import BeautifulSoup
requests.__path__
```

```python
# URLs for happiness by states from studies done in 2016, 2017, and 2018

url2018 = 'https://wallethub.com/edu/happiest-states/6959/#methodology'
url2017 = 'https://www.livescience.com/61744-happiest-states-2017-full-list.html'
url2016 = 'https://www.livescience.com/57716-happiest-states-2016-full-list.html'

response2018 = requests.get(url2018)
response2017 = requests.get(url2017)
response2016 = requests.get(url2016)
```

```python
# Take care of 2018 first

# convert to a BeautifulSoup Object

page2018 = response2018.text
soup2018 = BeautifulSoup(page2018, "lxml")
```

```python
# Iterate through 'td' tags on table and extract state name and happiness with a counter,
# putting them into a dict

happiness2018 = {}
count = 0
for link in soup2018.find_all('td'):
    trimmed_text = str(link.renderContents().strip())
#     print(count, trimmed_text)
    if (count + 5) % 6 == 0:
        state = trimmed_text[2:-1]
    if (count + 4) % 6 == 0:
        happiness2018[state] = float(trimmed_text[2:-1])

    count += 1

print(happiness2018)
```

```python
# Send dict to a df

happiness2018_df = pd.DataFrame.from_dict(happiness2018, orient='index', columns=['Happiness (2018)'])
happiness2018_df.head(50)
```

```python
# Pickle
happiness2018_df.to_pickle('~/ds/metis/metisgh/project_luther/data/happy2018.pkl')
```

```python
# 2017 next

# Scrape
page2017 = response2017.text
soup2017 = BeautifulSoup(page2017, "lxml")
```

```python
# Iterate through 'li' tags on table and extract state name and happiness with a counter,
# putting them into a dict

happiness2017 = {}
count = 0
for link in soup2017.find_all('li'):
    trimmed_text = str(link.renderContents().strip())
    if count >= 33 and count <= 82:
        info = trimmed_text[2:-1].split()
        if len(info) > 2:
            state = info[0] + ' ' + info[1]
        else:
            state = info[0]
        happiness2017[state] = float(info[-1])

    count += 1

print(happiness2017)
```

```python
# Send dict to a df
happiness2017_df = pd.DataFrame.from_dict(happiness2017, orient='index', columns=['Happiness (2017)'])
happiness2017_df.head(50)
```

```python
# Pickle
happiness2017_df.to_pickle('~/ds/metis/metisgh/project_luther/data/happy2017.pkl')
```

```python
# 2016 finally

# Scrape
page2016 = response2016.text
soup2016 = BeautifulSoup(page2016, "lxml")
```

```python
# Iterate through 'p' tags on table and extract state name and happiness with a counter,
# putting them into a dict

happiness2016 = {}
count = 0
for link in soup2016.find_all('p'):
    trimmed_text = str(link.renderContents().strip())
    if count >= 2 and count <= 51:
        info = trimmed_text[2:-1].split()
        if len(info) > 3:
            state = info[1] + ' ' + info[2]
        else:
            state = info[1]
        try:
            happiness2016[state] = float(info[-1])
        except ValueError:
            happiness2016[state] = float(info[-1][:-8])

    count += 1
```

```python
# Send dict to a df
happiness2016_df = pd.DataFrame.from_dict(happiness2016, orient='index', columns=['Happiness (2016)'])
happiness2016_df.head(50)
```

```python
# Pickle
happiness2016_df.to_pickle('~/ds/metis/metisgh/project_luther/data/happy2016.pkl')
```
