```python
# Import libraries for scraping

from __future__ import print_function, division
import requests
import re
import pandas as pd
from bs4 import BeautifulSoup
requests.__path__
```

```python
# Connect to website
url = 'https://www.currentresults.com/Weather/US/average-annual-state-temperatures.php'
response = requests.get(url)
```

```python
# Check the status
response.status_code # status code = 200 => OK
```

```python
# Scrape
page = response.text
soup = BeautifulSoup(page, "lxml")
```

```python
# Custom preprocessing for this webpage

# Iterate through 'td' tags on table and extract state name and temperature in fahrenheit with a counter,
# putting them into a dict

temp = {}
count = 0
for link in soup.find_all('td'):
    trimmed_text = str(link.renderContents().strip())
    if count % 4 == 0:
        state = re.search(">(.*)<", trimmed_text).group(1)
    if (count-1) % 4 == 0:
        temp[state] = float(trimmed_text[2:-1])

    count += 1

print(temp)
```

```python
# Send dict to a df
temp_df = pd.DataFrame.from_dict(temp, orient='index', columns=['Temperature (Fahrenheit)'])
temp_df.head(50)
```
