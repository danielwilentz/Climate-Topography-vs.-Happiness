```python
from __future__ import print_function, division
import requests
import re
import pandas as pd
from bs4 import BeautifulSoup
requests.__path__
```

```python
# URL for average annual state precipitation

url = 'https://www.currentresults.com/Weather/US/average-annual-state-precipitation.php'
response = requests.get(url)
```

```python
response.status_code # status code = 200 => OK
```

```python
# convert to a BeautifulSoup Object

page = response.text
soup = BeautifulSoup(page, "lxml")
```

```python
# Iterate through 'td' tags on table and extract state name and rainfall (in inches) with a counter,
# putting them into a dict

rainfall = {}
count = 0
for link in soup.find_all('td'):
    trimmed_text = str(link.renderContents().strip())
    if count % 4 == 0:
        state = re.search(">(.*)<", trimmed_text).group(1)
    if (count-1) % 4 == 0:
        rainfall[state] = float(trimmed_text[2:-1])

    count += 1

print(rainfall)
```

```python
# Send dict to a df

rainfall_df = pd.DataFrame.from_dict(rainfall, orient='index', columns=['Rainfall (Inches)'])
rainfall_df.head(50)
```
