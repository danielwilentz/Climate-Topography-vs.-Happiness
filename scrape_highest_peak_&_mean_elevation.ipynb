```python
from __future__ import print_function, division
import requests
import re
import pandas as pd
from bs4 import BeautifulSoup
requests.__path__
```

```python
# URL for highest peaks and mean elevation
url = 'https://en.wikipedia.org/wiki/List_of_U.S._states_and_territories_by_elevation'
response = requests.get(url)
```

```python
response.status_code # status code = 200 => OK
```

```python
# convert to a BeautifulSoup Object
page = response.text
soup = BeautifulSoup(page, "lxml")
```

```python
# Grab table from wikipedia page and iterate through table rows to get the table as a list
elevation_table = soup.find('table',{'class':'wikitable sortable'})
table_rows = elevation_table.findAll('tr')
table_as_list = []
for tr in table_rows:
    td = tr.find_all('td')
    row = [tr.text for tr in td]
    table_as_list.append(row)

table_as_list = table_as_list[1:]
```

```python
# Iterate through the table (as a list) to get state, highest peak, and mean elevation

states_list = []
highest_list = []
mean_list = []
for info in table_as_list:
    # extract state
    state = str(info[0]).strip()
    states_list.append(state)
    
    # extract highest peak
    highest = str(info[2]).strip()
    highest_list.append(float(highest.split('ft')[0].strip().replace(',', '')))
    
    # extract mean elevation
    mean = str(info[6]).strip()
    try:
        mean_list.append(float(mean.split('ft')[0].strip().replace(',', '')))
    except ValueError:
        mean_list.append(0)
    
# print(states_list)
# print(highest_list)
# print(mean_list)
```

```python
elevation_df = pd.DataFrame(
    {'State': states_list,
     'Highest Point': highest_list,
     'Mean Elevation': mean_list
    })
elevation_df.set_index('State', inplace = True)
```
